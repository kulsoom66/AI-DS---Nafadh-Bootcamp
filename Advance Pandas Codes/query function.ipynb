```python
import pandas as pd

# -------------------------------
# Create Sample DataFrame
# -------------------------------
df = pd.DataFrame({
    'X': range(5, 11),                # 6 values
    'Y': range(20, 140, 20),          # 6 values ✅ fixed
    'City': ['New York', 'Los Angeles', 'New York', 'Chicago', 'Los Angeles', 'Chicago']
})

print(df)


# -------------------------------
# 1. Basic AND condition
# -------------------------------
print("\n1. X < 8 AND City = New York")
print(df.query('(X < 8) & (City == "New York")'))


# -------------------------------
# 2. OR condition
# -------------------------------
print("\n2. City = New York OR Chicago")
print(df.query('(City == "New York") | (City == "Chicago")'))


# -------------------------------
# 3. NOT condition
# -------------------------------
print("\n3. NOT Los Angeles")
print(df.query('City != "Los Angeles"'))


# -------------------------------
# 4. Using variable inside query
# -------------------------------
city_name = "New York"
print("\n4. Using variable")
print(df.query('City == @city_name'))


# -------------------------------
# 5. Multiple variables
# -------------------------------
x_limit = 8
print("\n5. X < variable AND City = New York")
print(df.query('(X < @x_limit) & (City == "New York")'))


# -------------------------------
# 6. Using IN
# -------------------------------
print("\n6. City in list")
print(df.query('City in ["New York", "Chicago"]'))


# Using variable list
cities = ["New York", "Chicago"]
print("\n6b. City in variable list")
print(df.query('City in @cities'))


# -------------------------------
# 7. Range filtering
# -------------------------------
print("\n7. X between 6 and 9")
print(df.query('X >= 6 and X <= 9'))


# -------------------------------
# 8. String operations
# -------------------------------
print("\n8. City starts with 'New'")
print(df.query('City.str.startswith("New")'))

print("\n8b. City contains 'Angeles'")
print(df.query('City.str.contains("Angeles")'))


# -------------------------------
# 9. Create new column and filter
# -------------------------------
df['Z'] = df['X'] * df['Y']  # new calculated column
print("\n9. Z > 500")
print(df.query('Z > 500'))


# -------------------------------
# 10. Combine numeric + string
# -------------------------------
print("\n10. X > 6 AND City = Chicago")
print(df.query('(X > 6) & (City == "Chicago")'))


# -------------------------------
# 11. Sort after query
# -------------------------------
print("\n11. New York sorted by Y descending")
print(df.query('City == "New York"').sort_values(by='Y', ascending=False))


# -------------------------------
# 12. Select specific columns
# -------------------------------
print("\n12. Only X and City where X < 8")
print(df.query('X < 8')[['X', 'City']])


# -------------------------------
# 13. Query on index
# -------------------------------
df_indexed = df.set_index('City')  # set City as index
print("\n13. Query using index")
print(df_indexed.query('City == "New York"'))


# -------------------------------
# 14. Compare with normal filtering
# -------------------------------
print("\n14. Normal filtering vs Query")

print("\nNormal way:")
print(df[(df['X'] < 8) & (df['City'] == "New York")])

print("\nQuery way:")
print(df.query('(X < 8) & (City == "New York")'))
```
