```python
import pandas as pd

# Create two sample DataFrames
data1 = {
    'EmployeeID': [1, 2, 3, 4],
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Department': ['HR', 'IT', 'Finance', 'IT']
}

data2 = {
    'EmployeeID': [2, 4, 3, 5],
    'Salary': [70000, 80000, 60000, 90000]
}

# Convert dictionaries to DataFrames
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

# Display the original DataFrames
print("DataFrame 1:")
print(df1)
print("\nDataFrame 2:")
print(df2)

# Merge the DataFrames on 'EmployeeID' using an inner join
merged_df = pd.merge(df1, df2, on='EmployeeID', how='inner')
```

```python

labels = ['Low', 'Medium', 'High']

merged_df['SalaryCategory'] = pd.cut(merged_df['Salary'], bins=3, labels=labels)

print(merged_df)
```

```python
bins = [0, 65000, 75000, 90000]
labels = ['Low', 'Medium', 'High']

merged_df['SalaryCategory'] = pd.cut(merged_df['Salary'], bins=bins, labels=labels)

print(merged_df)
```
