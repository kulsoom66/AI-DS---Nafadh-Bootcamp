```python
import pandas as pd
import numpy as np

# Creating a more complex dataframe with multi-index
index = pd.MultiIndex.from_tuples([('New York', 'Q1', 'one'), ('New York', 'Q1', 'two'),
                                   ('Los Angeles', 'Q1', 'one'), ('Los Angeles', 'Q1', 'two'),
                                   ('Chicago', 'Q2', 'one'), ('Chicago', 'Q2', 'two'),
                                   ('New York', 'Q2', 'one'), ('New York', 'Q2', 'two')],
                                  names=['City', 'Quarter', 'ID'])

# Creating random data for Sales, Profit, and Quantity
sales = np.round(np.random.uniform(100, 500, 8), 2)  # Sales between 100 and 500
profit = np.round(np.random.uniform(0, 0.35, 8), 2)     # Profit between 0 and 0.35
quantity = np.random.randint(1, 20, 8)               # Quantity as integers

# Creating the dataframe
df_multi = pd.DataFrame({'Sales': sales, 'Profit(%)': profit, 'Quantity': quantity}, index=index)

print(df_multi)
```

```python
# ================================
# LEVEL 1
# ================================

# Q1: All data for New York
print("\nQ1")
print(df_multi.loc['New York'])

# Q2: New York in Q1
print("\nQ2")
print(df_multi.loc[('New York', 'Q1')])

# Q3: Los Angeles, Q1, ID two
print("\nQ3")
print(df_multi.loc[('Los Angeles', 'Q1', 'two')])

# Q4: Sales for Chicago
print("\nQ4")
print(df_multi.loc['Chicago', 'Sales'])

# Q5: Profit(%) for New York Q2
print("\nQ5")
print(df_multi.loc[('New York', 'Q2'), 'Profit(%)'])
```

```python
# ================================
# LEVEL 2
# ================================

# Q6: Sales & Quantity for New York Q1
print("\nQ6")
print(df_multi.loc[('New York', 'Q1'), ['Sales', 'Quantity']])

# Q7: Chicago Q2 → Sales only
print("\nQ7")
print(df_multi.loc[('Chicago', 'Q2'), 'Sales'])

# Q8: All rows where ID = 'one'
print("\nQ8")
print(df_multi.loc[(slice(None), slice(None), 'one'), :])

# Q9: All rows where Quarter = Q1
print("\nQ9")
print(df_multi.loc[(slice(None), 'Q1', slice(None)), :])
```

```python
# ================================
# LEVEL 4
# ================================

# Q14: New York → Quantity only
print("\nQ14")
print(df_multi.loc['New York', 'Quantity'])

# Q15: Quarter = Q2
print("\nQ15")
print(df_multi.loc[(slice(None), 'Q2', slice(None)), :])

# Q16: All cities, Q1, ID = one
print("\nQ16")
print(df_multi.loc[(slice(None), 'Q1', 'one'), :])

# Q17: Sales > 200
print("\nQ17")
print(df_multi[df_multi['Sales'] > 200])

# Q18: Quantity < 10 → only Quantity
print("\nQ18")
print(df_multi[df_multi['Quantity'] < 10][['Quantity']])
```
